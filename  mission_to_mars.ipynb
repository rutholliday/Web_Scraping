```python
url = 'https://mars.nasa.gov/news/8386/nasa-to-host-media-call-on-next-mars-landing-site/'
```

```python
# results are returned as an iterable list
results = soup.find_all('li', class_="slide")
title = soup.title.text
print(title)
```
